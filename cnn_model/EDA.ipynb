```python
df = pd.read_csv('../data/labels.csv')
```

```python
df['target'] = df['count']
```

```python
df[:1]
```

```python
df_sum = pd.DataFrame(df['target'].describe().round(1)).T
```

```python
fig, ax = plt.subplots(figsize=(8, 2)) # Adjust figure size as needed
ax.axis('off') # Hide axes
table = ax.table(cellText=df_sum.values, 
                 colLabels=df_sum.columns, 
                 loc='center',
                 cellLoc='center',
                )
table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1.3, 1.2) # Adjust cell size
plt.tight_layout()
plt.title('Target Variable Statistical Summary',
         fontsize=14,
          y=.8
         )
fig.patch.set_facecolor('whitesmoke')
ax.set_facecolor('ghostwhite')
plot_name = 'target_summary_table.png'
plt.savefig(plot_folder / plot_name, 
            bbox_inches='tight',
            pad_inches=.1,
            
           )
plt.show()
```

```python
data = df['count']
bin_width = 1
bins = np.arange(min(data), max(data) + bin_width, bin_width)
```

```python
bins
```

```python
fig, ax = plt.subplots(figsize=(7,4))
#df['count'].hist(bins='auto')
plt.hist(df['count'], 
         bins=bins, 
         color='cornflowerblue'
        )
ax.set_aspect('auto')
plt.title('Distribution of Target',
         fontsize=15
         )
plt.ylabel('Frequency (Number of Images)',
          fontsize=13
          )
plt.xlabel('Number of People (Target)',
          fontsize=14
          )
ax.tick_params(axis='both', 
               which='major', 
               labelsize=12
              ) 
plt.grid(ls=':', alpha=.6)
fig.patch.set_facecolor('lavender')
ax.set_facecolor('ghostwhite')
plot_name = 'target_histogram.png'
plt.savefig(plot_folder / plot_name, 
            bbox_inches='tight',
            pad_inches=.1,
            dpi=1000
            
           )
plt.show()
```
